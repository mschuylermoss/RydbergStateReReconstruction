# Plotting Function

```python
def movingaverage(values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
```

```python
def plot_all(delta_val,EXP_ds,QMC_ds,avg_fxn, exactE, RNN_type, x_ax=[None,None],y_ax=[None,None]):
    
    if RNN_type == 'OneD':
        nh=32
        lr = 0.0001
    else:
        nh=16
        lr = 0.001
    
    # Read in the data
    pathexp = f'./../data/N_256/Figure1/Exp_data/nh_{nh}/{RNN_type}_rnn/delta_'
    pathqmc = f'./../data/N_256/Figure1/QMC_data/dset_size_1000/nh_{nh}/{RNN_type}_rnn/delta_'
    pathqmc10k = f'./../data/N_256/Figure1/QMC_data/dset_size_10000/nh_{nh}/{RNN_type}_rnn/delta_'
    seed = '/seed_200'
    VMC = '/VMC_only'
    hybrid = '/hybrid_training/'
    ds = '_datasteps'
    lr = f'lr_{lr}/'
    Energy = '/Energy.npy'
    
    ExpData_train = np.load(pathexp+delta_val+seed+Energy)
    QMCData_train = np.load(pathqmc+delta_val+seed+Energy)
    QMCData10k_train = np.load(pathqmc10k+delta_val+seed+Energy)
    ExpH_train = np.load(pathexp+delta_val+seed+hybrid+lr+f'{EXP_ds}_datasteps'+Energy)
    QMCH_train = np.load(pathqmc+delta_val+seed+hybrid+lr+f'{QMC_ds}_datasteps'+Energy)
    
    ExpData_train = avg_fxn(ExpData_train,50)
    QMCData_train = avg_fxn(QMCData_train,50)
    QMCData10k_train = avg_fxn(QMCData10k_train,50)
    ExpH_train = avg_fxn(ExpH_train,50)
    QMCH_train = avg_fxn(QMCH_train,50)
    
    fig = plt.figure(1,figsize=(8,5), dpi=120, facecolor='w', edgecolor='k')
    greens = plt.get_cmap('Greens')
    plt.hlines(exactE,0,len(QMCData_train),label='QMC energy',linestyle='--',color='k')
    plt.plot(np.arange(len(QMCData_train)),QMCData_train,color='C3',label="QMC data")
    plt.plot(np.arange(len(ExpData_train)),ExpData_train,color='C1',label="Exp data")
    if QMCData10k_train.any() != None:
        plt.plot(np.arange(len(QMCData10k_train)),QMCData10k_train,color = 'C4',label = 'QMC data, 10k samples')
    plt.plot(np.arange(len(ExpH_train)),ExpH_train,color=greens(0.5),label=f'Exp Hybrid')
    plt.plot(np.arange(len(QMCH_train)),QMCH_train,color=greens(0.7),label=f'Exp Hybrid')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    x_ax = np.array(x_ax)
    y_ax = np.array(y_ax)
    if x_ax.any() != None:
        plt.xlim(x_ax[0],x_ax[1])
    if y_ax.any() != None: 
        plt.ylim(y_ax[0],y_ax[1])
    plt.title(f'delta = {delta_val}, {RNN_type} RNN Energy Convergence',fontsize=25)
    plt.ylabel('$\\langle H \\rangle$',fontsize=20)
    plt.xlabel('training steps',fontsize=20)
    plt.show()
```

# Get best energies per model

```python
def movingaverage(values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
```

```python
def GetBestEs1(delta,EXP_ds,QMC_ds,moving_avg_fxn,window):
    
    # Read in the data
    nh=32
    pathexp = f'./../data/N_256/Figure1/Exp_data/nh_{nh}/OneD_rnn/delta_'
    pathqmc = f'./../data/N_256/Figure1/QMC_data/dset_size_1000/nh_{nh}/OneD_rnn/delta_'
    pathqmc10k = f'./../data/N_256/Figure1/QMC_data/dset_size_10000/nh_{nh}/OneD_rnn/delta_'
    delta_s = str('%.3f' %delta)
    seed = '/seed_200'
    VMC = '/VMC_only'
    hybrid = '/hybrid_training/'
    ds = '_datasteps'
    lr = 'lr_0.0001/'
    Energy = '/Energy.npy'
    
    ExpData_train = np.load(pathexp+delta_s+seed+Energy)
    QMCData_train = np.load(pathqmc+delta_s+seed+Energy)
    QMCData10k_train = np.load(pathqmc10k+delta_s+seed+Energy)
    ExpH_train = np.load(pathexp+delta_s+seed+hybrid+lr+f'{EXP_ds}_datasteps'+Energy)
    QMCH_train = np.load(pathqmc+delta_s+seed+hybrid+lr+f'{QMC_ds}_datasteps'+Energy)

    # Take best of moving average for each model
    ExpData_E = np.min(moving_avg_fxn(ExpData_train,window))
    QMCData_E = np.min(moving_avg_fxn(QMCData_train,window))
    QMCData10k_E = np.min(moving_avg_fxn(QMCData10k_train,window))
    ExpH_E = np.min(moving_avg_fxn(ExpH_train,window))
    QMCH_E = np.min(moving_avg_fxn(QMCH_train,window))

    print('Exp: ',np.where(moving_avg_fxn(ExpData_train,window)==ExpData_E))
    print('QMC: ',np.where(moving_avg_fxn(QMCData_train,window)==QMCData_E))
    
#     plot_all(delta,QMCData_train,ExpData_train,ExpH_train,QMCH_train,qmcE_4455,'OneD',[None,None],[None,None],QMCData10k_train)
    
    return ExpData_E, QMCData_E, QMCData10k_E, ExpH_E, QMCH_E

def GetBestEs2(delta,moving_avg_fxn,window):
    
    # Read in the data
    nh = 16
    pathexp = f'./../data/N_256/Figure1/Exp_data/nh_{nh}/TwoD_rnn/delta_'
    pathqmc = f'./../data/N_256/Figure1/QMC_data/dset_size_1000/nh_{nh}/TwoD_rnn/delta_'
    pathqmc10k = f'./../data/N_256/Figure1/QMC_data/dset_size_10000/nh_{nh}/TwoD_rnn/delta_'
    delta_s = str('%.3f' %delta)
    seed = '/seed_200'
    VMC = '/VMC_only'
    hybrid = '/hybrid_training/'
    ds = '_datasteps'
    lr = 'lr_0.001/'
    Energy = '/Energy.npy'
    
    ExpData_train = np.load(pathexp+delta_s+seed+Energy)[0:2000]
    QMCData_train = np.load(pathqmc+delta_s+seed+Energy)[0:2000]
    QMCData10k_train = np.load(pathqmc10k+delta_s+seed+Energy)[0:2000]
    ExpH_train = np.load(pathexp+delta_s+seed+hybrid+lr+'50_datasteps'+Energy)
    QMCH_train = np.load(pathqmc+delta_s+seed+hybrid+lr+'50_datasteps'+Energy)

    # Take best of moving average for each model
    ExpData_E = np.min(moving_avg_fxn(ExpData_train,window))
    QMCData_E = np.min(moving_avg_fxn(QMCData_train,window))
    QMCData10k_E = np.min(moving_avg_fxn(QMCData10k_train,window))
    ExpH_E = np.min(moving_avg_fxn(ExpH_train,window))
    QMCH_E = np.min(moving_avg_fxn(QMCH_train,window))


    print('Exp: ',np.where(moving_avg_fxn(ExpData_train,window)==ExpData_E))
    print('QMC: ',np.where(moving_avg_fxn(QMCData_train,window)==QMCData_E))
    
    return ExpData_E, QMCData_E, QMCData10k_E, ExpH_E,QMCH_E
```

```python
ExpData_Es1 =[]
QMCData_Es1 =[]
QMCData10k_Es1 =[]
ExpData_Es2 =[]
QMCData_Es2 =[]
QMCData10k_Es2 =[]
ExpH_Es1 = []
ExpH_Es2 = []
QMCH_Es1 = []
QMCH_Es2 = []

deltas = np.array([-1.545,4.455,4.955,13.455])
exp_ds = np.array([600,4000,1400,1600])
qmc_ds = np.array([600,600,1000,5000])

for i in range(len(deltas)):
    delta_ = deltas[i]
    e_ds = exp_ds[i]
    q_ds = qmc_ds[i]
    print(delta_,e_ds,q_ds)
    print('OneD')
    ExpData_E1, QMCData_E1, QMCData10k_E1, ExpH_E1, QMCH_E1 = GetBestEs1(delta_,e_ds,q_ds,movingaverage,50)
    print('TwoD')
    ExpData_E2, QMCData_E2, QMCData10k_E2, ExpH_E2, QMCH_E2 = GetBestEs2(delta_,movingaverage,100)
    
    ExpData_Es1.append(ExpData_E1)
    QMCData_Es1.append(QMCData_E1)
    QMCData10k_Es1.append(QMCData10k_E1)
    ExpData_Es2.append(ExpData_E2)
    QMCData_Es2.append(QMCData_E2)
    QMCData10k_Es2.append(QMCData10k_E2)
    ExpH_Es1.append(ExpH_E1)
    ExpH_Es2.append(ExpH_E2)
    QMCH_Es1.append(QMCH_E1)
    QMCH_Es2.append(QMCH_E2)
```
