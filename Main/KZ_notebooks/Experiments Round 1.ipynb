```python
import numpy as np
import matplotlib.pylab as plt
```

# One D RNNs

### Train with data only

```python
oned_data_exp1_path = "data/N_256/delta_-1.545/OneD_rnn/OneD_data_lowdelta_long/"
oned_data_exp1_energy = np.load(oned_data_exp1_path+"Energy.npy")
oned_data_exp1_variance = np.load(oned_data_exp1_path+"Variance.npy")
oned_data_exp1_cost = np.load(oned_data_exp1_path+"Cost.npy")
```

```python
plt.plot(np.arange(len(oned_data_exp1_energy)),oned_data_exp1_energy)
plt.ylim(-1.2,-0.5)
plt.title("Energy")
plt.xlabel("step")
plt.show()
```

```python
plt.plot(np.arange(len(oned_data_exp1_variance)),oned_data_exp1_variance)
plt.ylim(0,30)
plt.title("Variance")
plt.xlabel("step")
plt.show()
```

```python
plt.plot(np.arange(len(oned_data_exp1_cost)),oned_data_exp1_cost)
plt.ylim(102,110)
plt.title("Cost")
plt.xlabel("step")
plt.show()
```

Remember that there is a constant in the KL that we ignore... taking that constant into account should mean that this cost is tending towards zero.
