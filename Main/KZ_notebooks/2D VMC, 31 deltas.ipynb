```python
def plot_all_es(array):
    fig = plt.figure(1,figsize=(8,5), dpi=120, facecolor='w', edgecolor='k')
    deltas,steps = np.shape(array)
    xaxis = np.arange(steps)
    delta = -1.545
    cmap = plt.get_cmap('Reds')
    for i in range(deltas):
        delta_str = str(delta)
        plt.plot(xaxis,array[i,:],label = delta_str,color=cmap(i/deltas))
        delta += 0.5
        delta = round(delta,3)
    plt.ylim(-6,0)
    plt.legend(ncol=2,loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xlabel('steps')
    plt.ylabel('$\\langle H \\rangle$')
    plt.title('Energy Convergence')
    plt.show()
    
def plot_all_vs(array):
    fig = plt.figure(1,figsize=(8,5), dpi=120, facecolor='w', edgecolor='k')
    deltas,steps = np.shape(array)
    xaxis = np.arange(steps)
    delta = -1.545
    cmap = plt.get_cmap('Blues')
    for i in range(deltas):
        delta_str = str(delta)
        plt.plot(xaxis,array[i,:],label = delta_str,color=cmap(i/deltas))
        delta += 0.5
        delta = round(delta,3)
#     plt.ylim(-0.1,10)
    plt.yscale('log')
    plt.legend(ncol=2,loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xlabel('steps')
    plt.ylabel('Var($\\langle H \\rangle$)')
    plt.title('Variance Convergence')
    plt.show()
```

```python
plot_all_es(energies)
plot_all_vs(variances)
```
