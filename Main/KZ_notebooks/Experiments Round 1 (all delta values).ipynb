```python
import numpy as np
import matplotlib.pylab as plt
```

# OneD RNNs

### VMC Only

```python
VMC_oned_lowdelta = "data/N_256/delta_-1.545/OneD_rnn/ns100/"
VMC_oned_lowdelta_e = np.load(VMC_oned_lowdelta+"Energy.npy")
VMC_oned_lowdelta_v = np.load(VMC_oned_lowdelta+"Variance.npy")
VMC_oned_lowdelta_c = np.load(VMC_oned_lowdelta+"Cost.npy")

VMC_oned_midlowdelta = "data/N_256/delta_4.455/OneD_rnn/ns100/"
VMC_oned_midlowdelta_e = np.load(VMC_oned_midlowdelta+"Energy.npy")
VMC_oned_midlowdelta_v = np.load(VMC_oned_midlowdelta+"Variance.npy")
VMC_oned_midlowdelta_c = np.load(VMC_oned_midlowdelta+"Cost.npy")

VMC_oned_midhighdelta = "data/N_256/delta_4.955/OneD_rnn/ns100/"
VMC_oned_midhighdelta_e = np.load(VMC_oned_midhighdelta+"Energy.npy")
VMC_oned_midhighdelta_v = np.load(VMC_oned_midhighdelta+"Variance.npy")
VMC_oned_midhighdelta_c = np.load(VMC_oned_midhighdelta+"Cost.npy")

VMC_oned_highdelta = "data/N_256/delta_13.455/OneD_rnn/ns100/"
VMC_oned_highdelta_e = np.load(VMC_oned_highdelta+"Energy.npy")
VMC_oned_highdelta_v = np.load(VMC_oned_highdelta+"Variance.npy")
VMC_oned_highdelta_c = np.load(VMC_oned_highdelta+"Cost.npy")
```

```python
fig = plt.figure(1,figsize=(8,5), dpi=120, facecolor='w', edgecolor='k')
plt.plot(np.arange(len(VMC_oned_lowdelta_e)),VMC_oned_lowdelta_e,label='delta = -1.545')
plt.plot(np.arange(len(VMC_oned_midlowdelta_e)),VMC_oned_midlowdelta_e,label='delta = 4.455')
plt.plot(np.arange(len(VMC_oned_midhighdelta_e)),VMC_oned_midhighdelta_e,label='delta = 4.955')
plt.plot(np.arange(len(VMC_oned_highdelta_e)),VMC_oned_highdelta_e,label='delta = 13.455')
plt.legend()
plt.ylim(-5,0)
plt.title('(OneD, VMC) Energy Convergence',fontsize=25)
plt.ylabel('$\\langle H \\rangle$',fontsize=20)
plt.xlabel('training steps',fontsize=20)
plt.show()
```

```python
fig = plt.figure(1,figsize=(8,5), dpi=120, facecolor='w', edgecolor='k')
plt.plot(np.arange(len(VMC_oned_lowdelta_v)),VMC_oned_lowdelta_v,label='delta = -1.545')
plt.plot(np.arange(len(VMC_oned_midlowdelta_v)),VMC_oned_midlowdelta_v,label='delta = 4.455')
plt.plot(np.arange(len(VMC_oned_midhighdelta_v)),VMC_oned_midhighdelta_v,label='delta = 4.955')
plt.plot(np.arange(len(VMC_oned_highdelta_v)),VMC_oned_highdelta_v,label='delta = 13.455')
plt.legend()
plt.ylim(-0.1,15)
plt.title('(OneD, VMC) Variance Convergence',fontsize=25)
plt.ylabel('Var($\\langle H \\rangle$)',fontsize=20)
plt.xlabel('training steps',fontsize=20)
plt.show()
```
