```python
deltas = data_16x16_15MHz_per_us['params'] * data_16x16_15MHz_per_us['rabi_freq']
print(deltas + data_16x16_15MHz_per_us['ryd_resonance'])
```

```python
print(deltas)
```

I think the above values are the physical values of delta that get controlled in the lab. I think the ryd_resonance is some intrinsic resonance that competes with the lab-controlled delta, so the deltas (based on values in params) are the "net delta" which is to say the delta applied in the lab plus the intrinsic resonance. (This is what they do in the Export_Raw_Data notebook)

The x-axis in the plots in the paper correspond to the values in params, so I am going to assume that these yield the values for delta that we should use as our Hamiltonian parameters (for VMC and QMC).
